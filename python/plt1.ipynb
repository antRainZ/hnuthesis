```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.font_manager import FontProperties
font_path = f'times+simsun.ttf'  # 替换为您自己的字体文件路径
font = FontProperties(fname=font_path)
```

```python
# 生成随机点的数据
x = np.random.uniform(1, 5, 1000)  # 在[1, 5]范围内生成1000个随机数作为横坐标
y = np.random.uniform(1, 1000, 1000)  # 在[1, 1000]范围内生成1000个随机数作为y轴的值

plt.yscale("log")

# 绘制散点图
plt.scatter(x, y)

# 添加标题、轴标签和图例（可选）
plt.title('散点图',font=font)
plt.xlabel('横轴',fontproperties=font)
plt.ylabel('纵轴',fontproperties=font)
plt.legend(['数据'],prop=font)

plt.savefig('../figures/data/testA.pdf',format="pdf")
# 显示图形
plt.show()
```

```python
# 构建数据
x = np.arange(1, 15)
y = [0.8, 0.88, 0.825, 0.76, 0.91, 0.95, 0.74, 0.99, 0.80, 0.72, 0.9, 0.81, 0.74, 0.87]
z = [37, 25, 17, 49, 27, 77, 34, 34, 34, 51, 39, 52, 47, 12]
u = [37, 31, 19, 57, 29, 86, 36, 37, 45, 64, 42, 57, 50, 24]
 
 
# 绘柱状图
plt.bar(x=x, height=z, label='实际', color='Coral', alpha=0.8)
plt.bar(x=x, height=u, label='总数', color='LemonChiffon', alpha=0.8)
# 在左侧显示图例
plt.legend(loc="upper left",prop=font)
 
# 设置标题
plt.title("Detection results",fontproperties=font)
# 为两条坐标轴设置名称
plt.xlabel("检测指标",fontproperties=font)
plt.ylabel("检测结果",fontproperties=font)
 
 
# 画折线图
ax2 = plt.twinx()
ax2.set_ylabel("召回率",fontproperties=font)
# 设置坐标轴范围
ax2.set_ylim([0.5, 1.05])
plt.plot(x, y, "r", marker='.', ms=5, linewidth='1', label="召回率")
# 显示数字
for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=8)
# 在右侧显示图例
plt.legend(loc="upper right",prop=font)
plt.savefig('../figures/data/testB.pdf',format="pdf")
 
plt.show()
```
